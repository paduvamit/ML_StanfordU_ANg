```octave
options = [' -t ' gaussian_kernel ' -g ' sigma ' -c ' C]
```

```octave
[cv_predicted_labels] = svmpredict(y_cv, X_cv, model);
```

```octave
cv_accuracy = mean(double(cv_predicted_labels == y_cv)) * 100
```
