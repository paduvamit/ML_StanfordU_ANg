```python
with open('examples') as f:
    examples = f.readlines()
```

```python
len(examples)
```

```python
from collections import defaultdict
```

```python
word_counts = defaultdict(int)
```

```python
for email in examples:
    for word in email.split():
        word_counts[word] += 1
```

```python
min_word_frequency = 100;
```

```python
frequent_words = sorted(list(filter(lambda x: x[1] >= min_word_frequency, word_counts.items())), key=lambda kv: kv[1], reverse=True)
```

```python
frequent_words = list(map(lambda x: x[0],frequent_words))
```

```python
len(frequent_words)
```

```python
frequent_words.sort()
```

```python
frequent_words
```

```python
f = open("spam_assassin_vocab.txt", "a")
for i in range(len(frequent_words)):
    line_delim = "" if i==(len(frequent_words)-1) else "\n";
    f.write(str(i+1) + "\t" + frequent_words[i] + line_delim)
f.close()
```
